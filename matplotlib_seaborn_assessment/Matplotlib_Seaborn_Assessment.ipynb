```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
iris= sns.load_dataset('iris')
iris
```

A. Creating and customizing line charts using Matplotlib


```python
sns.lineplot(data=iris)
```

B. Visualizing relationships between two or more variables using scatter plots

```python
sns.scatterplot(data=iris,x='sepal_length',y='petal_length')
```

C. Studying distributions of variables using histograms & bar charts to

```python
sns.histplot(data=iris)
```

```python
sns.barplot(data=iris)
```

D. Visualizing two-dimensional data using heatmaps


```python
sns.heatmap(iris.corr(), annot=True, fmt="f")
plt.show()
```

F. Plotting multiple Matplotlib and Seaborn charts in a grid

```python
col=['sepal_length','sepal_width','petal_length','petal_width']
fig, axs=plt.subplots(nrows=2,ncols=2,figsize=(20,10))
axs=axs.flatten()

for i, var in enumerate(col):
  sns.violinplot(x=var,data=iris,ax=axs[i])

fig.tight_layout()
plt.show()
```
